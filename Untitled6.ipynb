```python
#linear regression

import pandas as pd
a=pd.read_csv('reg.csv')
print(a)
a.duplicated().any()
a['gives_birth'].fillna(1,inplace=True)
a['aquatic_animal'].fillna(0,inplace=True)
x=a.iloc[:,1:5].values
#print(x)
y=a.iloc[:,5].values
#print(y)

from sklearn.preprocessing import LabelEncoder as ll
r=ll()
x[7:,0]=r.fit_transform(x[7:,0])
x[7:,1]=r.fit_transform(x[7:,1])
x[:,3]=r.fit_transform(x[:,3])
print(x)
y=r.fit_transform(y)
print(y)
import numpy as np
x=x[:,np.newaxis,2]
print(x)

from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.25,random_state=0)

from sklearn.linear_model import LinearRegression as rr
c=rr()
c.fit(x_train,y_train)
p=c.predict(x_test)
print(p)

from sklearn.metrics import mean_squared_error as ms,r2_score as rs
a=ms(y_test,p)
print(a)
b=rs(y_test,p)
print(b)

x_test=x_test.astype('i')
print(x_test)
y_test=y_test.astype('i')
print(y_test)
from matplotlib import pyplot as plt

plt.scatter(x_test,y_test,color="y")
plt.plot(x_test,p,color="g")
plt.xlabel("aquatic_animal")
plt.ylabel("class_label")
```

```python
#kmean clustering

import pandas as pd
a=pd.read_csv('kmea.csv')
print(a)

a['gives_birth'].fillna(1,inplace=True)
a['aquatic_animal'].fillna(0,inplace=True)
print(a)

x=a.iloc[:,1:].values
print(x)

from sklearn.preprocessing import LabelEncoder as ll
r=ll()
x[7:,0]=r.fit_transform(x[7:,0])
x[7:,1]=r.fit_transform(x[7:,1])
print(x)

from sklearn.cluster import KMeans
kmean=KMeans(init = 'k-means++',n_clusters=2,random_state=0).fit(x)
p=kmean.predict(x)
print(p)

x=x.astype('i')
print(x)

import matplotlib.pyplot as plt
colors=list(map(lambda x:'red' if x==1 else 'blue',p))
plt.scatter(x[:,0],x[:,1],c=colors)
```

```python
#knn

import pandas as pd
a=pd.read_csv('kn.csv')
print(a.to_string())
a.duplicated().any()
a.drop_duplicates(inplace=True)
#print(a)

m=a['gives_birth'].mean()
print(m)
a['gives_birth'].fillna(m,inplace=True)
#print(a)

x=a.iloc[:,1:5].values
y=a.iloc[:,5].values

#print(y)

from sklearn.preprocessing import LabelEncoder as ll
r=ll()
x[:,1]=r.fit_transform(x[:,1])
x[:,2]=r.fit_transform(x[:,2])
x[:,3]=r.fit_transform(x[:,3])
#print(x)

y=r.fit_transform(y)
#print(y)

from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.25,random_state=0)

from sklearn.neighbors import KNeighborsClassifier
c=KNeighborsClassifier()
c.fit(x_train,y_train)
p=c.predict(x_test)
print("x_test\n",p)

from sklearn.metrics import accuracy_score,classification_report,confusion_matrix
print(accuracy_score(y_test,p))
```

```python
#naive bayes

import pandas as pd
data=pd.read_csv('bayes.csv')


data.duplicated().any()
data.drop_duplicates(inplace=True)
print("\n",data.to_string())

mean_b=data['gives_birth'].mean()
#print(mean_b)
mean_aq=data['aquatic_animal'].mean()
#print(mean_aq)

data['gives_birth'].fillna(mean_b,inplace=True)
data['aquatic_animal'].fillna(mean_aq,inplace=True)

x=data.iloc[:,1:5].values
#print(x)
y=data.iloc[:,5:].values
#print(y)

from sklearn.preprocessing import LabelEncoder as ll
r=ll()
y=r.fit_transform(y)
print(y)

from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.25,random_state=0)

from sklearn.naive_bayes import GaussianNB
c=GaussianNB()
c.fit(x_train,y_train)
p=c.predict(x_test)
print(p)

from sklearn.metrics import accuracy_score,classification_report,confusion_matrix
print(confusion_matrix(y_test,p))
print(accuracy_score(y_test,p))
print(classification_report(y_test,p))
```

```python
#decision tree


import pandas as pd
a=pd.read_csv('kn.csv')
print(a.to_string())
a.duplicated().any()
a.drop_duplicates(inplace=True)
#print(a)

m=a['gives_birth'].mean()
print(m)
a['gives_birth'].fillna(m,inplace=True)
#print(a)

x=a.iloc[:,1:5].values
y=a.iloc[:,5].values

#print(y)

from sklearn.preprocessing import LabelEncoder as ll
r=ll()
x[:,1]=r.fit_transform(x[:,1])
x[:,2]=r.fit_transform(x[:,2])
x[:,3]=r.fit_transform(x[:,3])
#print(x)

#y=r.fit_transform(y)
#print(y)

from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.25,random_state=0)


from sklearn.tree import DecisionTreeClassifier as D
from sklearn import tree
import matplotlib.pyplot as plt 

d=D(criterion='entropy')
d.fit(x_train,y_train)
plt.figure(figsize=(20,20))
tree.plot_tree(d)
plt.show()

p=d.predict(x_test)
print(p)
```
