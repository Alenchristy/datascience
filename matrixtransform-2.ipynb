```python
#matrix transformation

from PIL import Image
import numpy as np
from scipy import ndimage
from matplotlib import pyplot as plt
%matplotlib inline
a=Image.open('batm.jpg')
b=np.array(a)
c=ndimage.rotate(b,260,cval=110)
d=Image.fromarray(c)
d.save('bt.jpg')
plt.imshow(d)
```

```python
#scaling
from PIL import Image
import numpy as np
from scipy import ndimage
from matplotlib import pyplot as plt
%matplotlib inline
a=Image.open('batm.jpg')
b=np.array(a)
c=ndimage.zoom(b,(0.25,0.1,1))
d=Image.fromarray(c)
d.save('bt1.jpg')
plt.imshow(d)
```

```python
#shearing

from PIL import Image
import numpy as np
from scipy import ndimage
from matplotlib import pyplot as plt
%matplotlib inline
a=Image.open('batm.jpg')
b=np.array(a)
h,w,c=b.shape
d=np.array([[1,0.5,0],[0.5,1,0],[0,0,1]])
e=ndimage.affine_transform(b,d)
f=Image.fromarray(e)
f.save('bt2.jpg')
plt.imshow(f)
```
