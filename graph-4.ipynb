```python
#programs using matplotlib / plotly / bokeh / seabron for data visualisation.
import numpy as np
import matplotlib.pyplot as plt

x=np.array([1,3,6,8])
y=np.array([3,5,7,11])
plt.plot(x,y,ms=10,linestyle='dashed',color='g',marker='*')
plt.title('graph',loc='center')

plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

x=np.array([1,3,6,8])
y=np.array([1,3,5,7])
plt.bar(x,y)
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

x=np.array([1,3,6,8])
y=np.array([1,3,6,8])
plt.hist(x,y,width=2)
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
x=np.array([100,100,160])
y=([0,0.2,0])
a=["apple","orange","mango"]
plt.pie(x,labels=a,explode=y,shadow=True)
plt.legend(loc='right')
plt.show()
```
